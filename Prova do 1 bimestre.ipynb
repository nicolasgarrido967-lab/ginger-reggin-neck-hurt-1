ATIVIDADE 1 - Par ou ímpar
Peça um número ao usuário e informe se ele é par ou ímpar.


```python
numero = int(input("digite um numero:"))

if numero % 2 == 0:
  print("o numero é par")
else:
  print("o numero é impar")
```

ATIVIDADE 2 - Maior de dois números
Solicite dois números e mostre qual é o maior (ou se são iguais).

```python
n1 = int(input("digite o primeiro numero"))
n2 = int(input("digite o segundo numero"))

if n1 > n2:
  print("o primeiro numero é maior que o segundo")
if n1 < n2:
  print("o segundo numero é maior que o primeiro")
if n1 == n2:
  print("os numeros são iguais")
```

ATIVIDADE 3 - Aprovação de aluno
Peça a nota de um aluno e informe:
Aprovado (nota ≥ 7)
Recuperação (5 ≤ nota < 7)
Reprovado (nota < 5)


```python
n1 = int(input("digite a primeira nota"))
n2 = int(input("digite a segunda nota"))
n3 = int(input("digite a terceira nota"))
n4 = int(input("digite a quarta nota"))

media = (n1 + n2 + n3 + n4) / 4
if media > 10 or n1 > 10 or n2 > 10 or n3 > 10 or n4 > 10:
  print("digite numeros validos")

if media >= 7:
  print(f"você foi aprovado com uma média de {media}")
if media >= 5 and media < 7:
  print(f"você ficou de recuperação com uma média de {media}")
if media < 5:
  print(f"você foi reprovado com uma média de {media}")
```

ATIVIDADE 4 - Classificação de idade
Receba a idade e classifique:
Criança (0–12)
Adolescente (13–17)
Adulto (18–59)
Idoso (60+)


```python
idade = int(input("digite sua idade:"))
if idade <= 12:
  print("você é uma criança")
if idade >= 13 and idade <= 17:
  print("você é um adolescente")
if idade >= 18 and idade <= 59:
  print("você é um adulto")
if idade > 60:
  print("você é um idoso")
```

ATIVIDADE 6 - Positivo, negativo ou zero
Peça um número e diga se ele é positivo, negativo ou zero. Positivo, negativo ou zero

ATIVIDADE 7 - Calculadora simples
Peça dois números e uma operação (+, -, *, /). Use if/elif para calcular o resultado.

```python
num1 = float(input("Digite o primeiro número: "))
num2 = float(input("Digite o segundo número: "))
operacao = input("Digite a operação (+, -, *, /): ")

if operacao == '+':
  resultado = num1 + num2
elif operacao == '-':
  resultado = num1 - num2
elif operacao == '*':
  resultado = num1 * num2
elif operacao == '/':
  if num2 != 0:
    resultado = num1 / num2
  else:
    resultado = "Erro: Divisão por zero!"
else:
  resultado = "Operação inválida!"

print(f"Resultado: {resultado}")
```

ATIVIDADE 8 - Dia da semana
Peça um número de 1 a 7 e mostre o dia correspondente (ex: 1 = domingo).

```python
numero_dia = int(input("Digite um número de 1 a 7:"))

if numero_dia == 1:
  print("Domingo")
elif numero_dia == 2:
  print("Segunda-feira")
elif numero_dia == 3:
  print("Terça-feira")
elif numero_dia == 4:
  print("Quarta-feira")
elif numero_dia == 5:
  print("Quinta-feira")
elif numero_dia == 6:
  print("Sexta-feira")
elif numero_dia == 7:
  print("Sábado")
else:
  print("Número inválido. Digite um número de 1 a 7.")
```

ATIVIDADE 9 - Ano bissexto
Solicite um ano e informe se ele é bissexto ou não.

```python
ano = int(input("Digite um ano:"))

if (ano % 4 == 0 and ano % 100 != 0) or (ano % 400 == 0):
  print(f"O ano {ano} é bissexto.")
else:
  print(f"O ano {ano} não é bissexto.")
```
