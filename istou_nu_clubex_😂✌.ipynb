# RECEBER UM PROGRAMA QUE RECEBE 04 NOTAS DO ALUNO E MOSTRA SE O ALUNO FOI APROVADO

```python
# ENTRADA

nota1 = float(input('Digite a primeira nota '))
nota2 = float(input('Digite a segunda nota '))
nota3 = float(input('Digite a terceira nota '))
nota4 = float(input('Digite a quarta nota '))

# PROCESSAMENTO

media = (nota1 + nota2 + nota3 + nota4) / 4

if media >= 7:
 print(f"O aluno foi aprovado com média {media}")
elif media >= 5:
 print(f"O aluno está de recuperação com média {media}")
else:
 print(f"O aluno foi reprovado com nota média {media}")
```
