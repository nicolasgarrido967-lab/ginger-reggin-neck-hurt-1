```python
#operdores aritimétricos

#fazer um programa que o usuario digit dois numeros
#saída mostrar operação aritimétrica


#entrada

a = int(input("Digite o primeiro numero: "))
b = int(input("Digite o segundo numero: "))

#processamento

soma = a + b
subtracao = a - b
multiplicacao = a * b
divisao = a / b
div_inteira = a // b
resto = a % b
potencia = a ** b


#saída

print (f"o resultado da soma de {a} mais {b} é igual a: {soma}")
print (f"o resultado da subtacao de {a} por {b} é igual a: {subtracao}")
print (f"o resultado da multiplicação de {a} por {b} é igual a: {multiplicacao}")
print (f"o resultado de divisao de {a} por {b} é igual a: {divisao}")
print (f"o resultado da div_inteira de {a} por {b} é igual a: {div_inteira}")
print (f"o resultado do resto de {a} por {b} é igual a: {resto}")
print (f"o resultado da potencia de {a} por {b} é igual a: {potencia}")
```

# FAZER UM PROGRAMA QUE CONVERTE HORAS EM MINUTOS E HORAS EM SEGUNDOS

```python
# ENTRADA

horas = int(input("Digite a quantidade de horas: "))

# PROCESSAMENTO
horas_em_minutos = horas * 60
horas_em_segundos = horas * 3600


# SAÍDA
print(f"{horas} horas equivalem a {horas_em_minutos} minutos")
print(f"{horas} horas equivalem a {horas_em_segundos} segundos")

```
