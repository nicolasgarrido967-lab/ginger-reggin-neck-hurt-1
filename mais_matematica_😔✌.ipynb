criar um programa que verifica se o numero épar ou impar

```python
#entrada

numero = int(input("dgite um numero"))

#processamento

if numero % 2 == 0:
  print("o numero é par")
else:
  print("o numero é impar")
```

```python
#escrever um programa que o usuario digita o nome, recebe 4 notas referente aos bimestres, calcula a média e mostra se o aluno foi aprovado, recuperação aprovado

#entrada

nome = input("digite seu nome:")
nota1 = int(input("digite a primeira nota"))
nota2 = int(input("digite a segunda nota"))
nota3 = int(input("digite a terceira nota"))
nota4 = int(input("digite a quarta nota"))
media = (nota1 + nota2 + nota3 + nota4) / 4
if nota1 > 10 or nota2 > 10 or nota3 > 10 or nota4 > 10:
  print(f"digite uma nota valida bro :/")
#processamento

if media >= 7:
  print(f"parabens {nome}, você foi aprovado com uma media de {media}")
if media >= 5 and media < 7:
  print(f"{nome}, você ficou de recuperação com uma média de {media}")
if media < 5:
  print(f"{nome}, você foi reprovado com uma média de {media}")
```

receber 3 medidas de lados, verificar se é triangulo valido e classificação

```python
lado1 = float(input("digite o primeiro lado: "))
lado2 = float(input("digite o segundo lado: "))
lado3 = float(input("digite o terceiro lado: "))


if (lado1 + lado2 > lado3) and (lado1 + lado3 > lado2) and (lado2 + lado3 > lado1):
    print("É um triângulo válido!")

    if lado1 == lado2 and lado2 == lado3:
        print("É um triângulo equilátero.")
    elif lado1 == lado2 or lado1 == lado3 or lado2 == lado3:
        print("É um triângulo isósceles.")
    else:
        print("É um triângulo escaleno.")
else:
    print("Não é um triângulo válido.")
```
